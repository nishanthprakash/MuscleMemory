1. Median of two sorted arrays  
    Paul's solution:
    Iterate until left with one element:
      - Partition the two arrays into their halves -- A1, A2 and B1, B2 
      - Last elements of A1/B1 or first elements of A2/B2 define median ranges. Find if the median lies in B2, A1 or A2, B1 
      - throw away the minimum number of elements of the intervals of A and B where the median doesn't lie
      - Handle these base cases separately:
            The smaller array has only one element
                Case 0: N = 0, M = 2
                Case 1: N = 1, M = 1.
                Case 2: N = 1, M is odd
                Case 3: N = 1, M is even
            The smaller array has only two elements
                Case 4: N = 2, M = 2
                Case 5: N = 2, M is odd
                Case 6: N = 2, M is even

    Runs in $\log_2(min(n, m))$ time

5. Regex Matching (different from wildcard matching)
    match(pat, i , str, j) 
    - i==len(pat) and j==len(str) => True
    - i==len(pat) or j==len(str) => False
    - pat[i]==str[j] => match(pat, i+1 , str, j+1)
    - pat[i]=='.' and guarded(pat[i+1])=='\*' =>  any(match(pat, i+1 , str, k) for k in range(j, len(str)))
    - pat[i]=='.' => match(pat, i+1 , str, j+1)
    - pat[i]=='\*' => {assert pat[i-1] != '.'} any(while pat[i-1] == str[j++]: match(pat, i+1 , str, j+1))
    easily bottom up DP done like wildcard matching

7. Range Minimum Query (RMQ) -- useful for lowest common ancestor, longest common prefix  
    brute => O(n^2) time  
    preprocessed range-min matrix => O(n^2) space, O(n^2) preprocesstime  
    Segment trees => O(n) preprocesstime, O(n) space, O(logn) querytime  
       Preprocess by storing min(lefttree, righttree) in heap like array
       Overlaps with req. interval: partial => recur, none=> return +INF, full=> return curr.min_augmented
   Sparse Table => O(nlogn) space and O(1) time
       Store in n*(lgn+1) matrix at i, j, the index of min element from i to i+2^j (fill column-wise), i to 2^j is min of i to i+2^j/2 and i+2^j/2 to i+2^j/2+2^j/2 (previously calc.) => O(nlogn preprocess)
       In the query i to k, get min of j = log(k-i+1), i to i+2^j (pre-calculated) and i+(k-i+1 - 2^j) to next 2^j (pre-calculated)

20. Max-flow
    - Ford&Fulkerson: find a path (whatever-First-Search), find the bottleneck capacity and augment it. Time complexity O(f\*E), f being flow capacity and WFS requiring O(E) (assert: all edge capacities have to rational for halting)
    - Edmonds&Karp: choose the next shortest augmenting path from the source to the sink (BFS)  
        Jeff says pick the max bottleneck path (emulate Kruskal's algo for MST from src to sink)
    - Dinits: choose augmenting path with fewest edges
        import decimal

            def EdmondsKarp(E, C, s, t):
                n = len(C)
                flow = 0
                F = [[0 for y in range(n)] for x in range(n)]
                while True:
                    P = [-1 for x in range(n)]
                    P[s] = -2
                    M = [0 for x in range(n)]
                    M[s] = decimal.Decimal('Infinity')
                    BFSq = []
                    BFSq.append(s)
                    pathFlow, P = BFSEK(E, C, s, t, F, P, M, BFSq)
                    if pathFlow == 0:
                        break
                    flow = flow + pathFlow
                    v = t
                    while v != s:
                        u = P[v]
                        F[u][v] = F[u][v] + pathFlow
                        F[v][u] = F[v][u] - pathFlow
                        v = u
                return flow

            def BFSEK(E, C, s, t, F, P, M, BFSq):
                while (len(BFSq) > 0):
                    u = BFSq.pop(0)
                    for v in E[u]:
                        if C[u][v] - F[u][v] > 0 and P[v] == -1:
                            P[v] = u
                            M[v] = min(M[u], C[u][v] - F[u][v])
                            if v != t:
                                BFSq.append(v)
                            else:
                                return M[t], P
                return 0, P
